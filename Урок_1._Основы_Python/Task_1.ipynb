### Решение задачи 1.2

```python
# Вариант 1:

# Запрашиваем год у пользователя
year = int(input("Введите год: "))

# Проверяем, является ли год високосным
if year % 4 == 0:
    # Год високосный, если он кратен 4
    if year % 100 == 0:
        # Но не високосный, если кратен 100 и не кратен 400
        if year % 400 != 0:
            print(f"{year} - невисокосный год")
        else:
            print(f"{year} - високосный год")
    else:
        print(f"{year} - високосный год")
else:
    print(f"{year} - невисокосный год")
```

## Этот код используется для определения, является ли введенный год високосным или нет. Вот подробное объяснение каждой части кода для варианта 1 задачи 1.2:

3. `if year % 100 == 0:` - Этот условный оператор проверяет, делится ли введенный год на 100 без остатка. Если это так, то выполняется следующий блок кода.

4. `if year % 400 != 0:` - Этот условный оператор проверяет, не делится ли введенный год на 400 без остатка. Если это так, то год считается невисокосным.

6. `else:` - Это ключевое слово используется для обозначения альтернативного блока кода, который будет выполнен, если предыдущее условие не выполнено.

8. `else:` - Этот блок кода выполняется, если ни одно из предыдущих условий не выполнено, что означает, что введенный год не является високосным.

9. `print(f"{year} - невисокосный год")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год не является високосным.

## Этот код используется для определения, является ли введенный год високосным или нет. Вот подробное объяснение каждой части кода для варианта 2 задачи 1.2:

1. `year = int(input("Введите год: "))` - Эта строка кода просит пользователя ввести год и сохраняет его в переменной `year`.

2. `print(f"{year} - {('високосный' if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) else 'невисокосный')} год")` - Эта строка выполняет две вещи:

   - Использует условный оператор `if` для определения статуса года.

   Условный оператор `if` в данном случае работает следующим образом:
   - Сначала проверяется, делится ли год на 4 без остатка (`year % 4 == 0`). Если это так, то условие истинно.

   В конце строки используется форматная строка `f"{year} - {('високосный' if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) else 'невисокосный')} год"` для форматирования вывода.

## Задача 2

- Пользователь вводит число от 1 до 999.

- Используя операции с числами сообщите что введено: однозначное число, двузначное чисоло или трехзначное число.

- Для однозначного числа верните её квадрат, например 5 = 25

- Для двухзначного числа произведение цифр, например 30 = 3

- Для техзначного чила его зеркальное отображение, например 520 = 25

- Если число не издиапазона, запросите новое число.

- В коде должен быть один input и один print.

```python
# вариант 1

# Считываем число от пользователя
number = int(input("Введите число от 1 до 999: "))

# Проверяем, что число находится в диапазоне от 1 до 999
while number < 1 or number > 999:
    print("Число должно быть от 1 до 999")
    number = int(input("Введите число от 1 до 999: "))

# Определяем количество цифр в числе и выполняем соответствующую операцию
if number < 10:
    # Однозначное число
    print(number   *   number)  # Выводим квадрат числа
elif number < 100:
    # Двузначное число
    first_digit = number // 10  # Получаем первую цифру числа
    second_digit = number % 10  # Получаем вторую цифру числа
    print(first_digit   *   second_digit)  # Выводим произведение цифр
else:
    # Трехзначное число
    reversed_number = str(number)[::-1]  # Зеркальное отображение числа
    print(int(reversed_number))  # Выводим зеркальное отображение числа
```

## Подробное объяснение каждой части кода для задачи 2 вариант 1:

Этот код представляет собой программу, которая запрашивает у пользователя число в диапазоне от 1 до 999. После ввода числа программа проверяет, находится ли оно в указанном диапазоне. Если нет, то программа сообщает об ошибке и снова запрашивает число.

Если число находится в диапазоне, программа определяет, является ли оно однозначным, двузначным или трехзначным, и выполняет соответствующую операцию: для однозначного числа вычисляет и выводит его квадрат, для двузначного числа вычисляет и выводит произведение его цифр, а для трехзначного числа выводит его зеркальное отображение.

Вот подробное объяснение кода:

1. Сначала программа считывает число от пользователя с помощью функции `input()`.
2. Затем она проверяет, находится ли введенное число в диапазоне от 1 до 999, используя цикл `while`. Если число вне диапазона, программа выводит сообщение об ошибке и запрашивает новое число.
3. Если число находится в диапазоне, программа проверяет, сколько цифр в числе, используя условные операторы `if` и `elif`.
4. Для однозначного числа программа вычисляет его квадрат и выводит результат.
5. Для двузначного числа программа делит число на `10`, чтобы получить первую цифру, и использует оператор остатка `%` для получения второй цифры. Затем она умножает эти две цифры и выводит результат.
6. Для трехзначного числа программа преобразует число в строку, а затем использует обратную индексацию `[::-1]` для получения зеркального отображения числа. После этого она преобразует полученную строку обратно в число и выводит результат.

```python
# Вариант 2

# Запрос числа от пользователя
number = int(input("Введите число от 1 до 999: "))

# Переменная для хранения результата
result = ""

# Проверка числа и выполнение соответствующих операций
if 1 <= number <= 9:
    result = f"Однозначное число, его квадрат: {number ** 2}"
elif 10 <= number <= 99:
    result = f"Двузначное число, произведение его цифр: {(number // 10) * (number % 10)}"
elif 100 <= number <= 999:
    result = f"Трехзначное число, его зеркальное отображение: {int(str(number)[::-1])}"
else:
    result = "Число не входит в заданный диапазон от 1 до 999."

# Вывод результата
print(result)
```

### Подробное объяснение каждой части кода для задачи 2 вариант 2:

Этот код предназначен для выполнения различных математических операций в зависимости от введенного числа. Пользователь вводит число от 1 до 999, после чего программа выполняет одну из трех операций:

1. Если число однозначное (от 1 до 9), то вычисляется его квадрат.
2. Если число двузначное (от 10 до 99), то вычисляется произведение его цифр.
3. Если число трехзначное (от 100 до 999), то вычисляется его зеркальное отображение (число, полученное путем чтения его справа налево).

Если введенное число не попадает в диапазон от 1 до 999, программа выводит сообщение об ошибке.

Вот как работает каждая часть кода:

1. `number` = `int(input("Введите число от 1 до 999: "))` - Эта строка кода считывает ввод пользователя и преобразует его в целое число.

2. `result = ""` - Создается пустая переменная `result`, которая будет использоваться для хранения результата операции.

3. `if 1 <= number <= 9:` - Этот блок кода выполняется, если введенное число находится в диапазоне от 1 до 9. В этом случае вычисляется квадрат числа и результат сохраняется в переменной `result`.

4. `elif 10 <= number <= 99:` - Этот блок кода выполняется, если введенное число находится в диапазоне от 10 до 99. В этом случае вычисляется произведение цифр числа и результат сохраняется в переменной `result`.

5. `elif 100 <= number <= 999:` - Этот блок кода выполняется, если введенное число находится в диапазоне от 100 до 999. В этом случае вычисляется зеркальное отображение числа и результат сохраняется в переменной `result`.

6. `else:` - Этот блок кода выполняется, если ни одно из предыдущих условий не выполнено, что означает, что введенное число не попадает в заданный диапазон. В этом случае в переменную result записывается сообщение об ошибке.

7. `print(result)` - В конце кода выводится результат операции или сообщение об ошибке.

## Задача 3

Вывести в консоль таблицу умножения

от 2*2 

до 9*9

Как на школьной тетрадки

```python
# вариант 1

for i in range(2, 10):
    for j in range(2, 10):
        print(f"{i} * {j} = {i * j}", end="\t")  # \t добавляет табуляцию для выравнивания столбцов
    print("")  # Печать новой строки после каждой строки таблицы умножения
```

### Подробное объяснение каждой части кода задачи 3 вариант 1:

Этот код создает таблицу умножения чисел от 2 до 9.

1. `for i in range(2, 10):` Это начало внешнего цикла `for`. Цикл будет выполняться для каждого значения переменной `i`, начиная с 2 и заканчивая 9.

2. `for j in range(2, 10):` Это начало внутреннего цикла `for`. Он будет выполняться для каждого значения переменной `j`, также начиная с 2 и заканчивая 9.

3. `print(f"{i}  *  {j} = {i  *  j}", end="\t")` - Эта строка печатает результат умножения `i` на `j`, используя форматирование строк. Значение `end="\t"` указывает, что после вывода результата не нужно добавлять перевод строки, а вместо этого использовать символ табуляции `\t`, чтобы выровнять результаты в столбцы.

4. `print("")` - Эта строка печатает пустую строку, чтобы начать новую строку после завершения внутреннего цикла `for`.

Таким образом, программа выведет все возможные произведения чисел от 2 до 9, выровненные в виде таблицы умножения.

```python
# вариант 2

# Создаем список для хранения результатов умножения
results = []

# Перебираем числа от 2 до 9
for i in range(2, 10):
    # Создаем новый список для хранения результатов умножения текущего числа
    current_results = []
    # Перебираем числа от 2 до 9
    for j in range(2, 10):
        # Вычисляем произведение двух чисел и добавляем его в текущий список результатов
        current_results.append(i  *  j)
    # Добавляем текущий список результатов в общий список
    results.append(current_results)

# Выводим результаты в консоль
for row in results:
    print(' '.join(map(str, row)))
```

### Подробное объяснение каждой части кода задачи 3 вариант 2:

Этот код создает таблицу умножения чисел от 2 до 9. 

1. `results = []:` Создаем пустой список, который будет использоваться для хранения всех результатов умножения.

2. `for i in range(2, 10):` Начинаем внешний цикл `for`, который будет перебирать числа от 2 до 9.

3. `current_results = []:` Внутри внешнего цикла создаем пустой список `current_results`, который будет хранить результаты умножения текущего числа `i`.

4. `for j in range(2, 10):` Начинаем внутренний цикл `for`, который будет перебирать числа от 2 до 9 для каждого значения `i`.

5. `current_results.append(i  *  j):` В каждом цикле внутреннего цикла умножаем текущее значение `i` на текущее значение `j` и добавляем результат в список `current_results`.
