## Задача 1

Напишите программу,которая запрашивает год и проверяет его на високосность.

Распишите все возможные проверки в цепочке `elif`

Обязательно учтите год ввода по Григорианскому календарю 

### Решение задачи 1

```python
# Вариант 1:

# Запрос года у пользователя
year = int(input("Введите год для проверки: "))

# Проверяем, является ли год високосным
if year < 1582:
    print(f"Год {year} не соответствует Григорианскому календарю.")
elif year % 400 == 0:
    print(f"Год {year} является високосным.")
elif year % 100 == 0:
    print(f"Год {year} не является високосным.")
elif year % 4 == 0:
    print(f"Год {year} является високосным.")
else:
    print(f"Год {year} не является високосным.")
```

### Этот код используется для определения, является ли введенный год високосным или нет. Вот подробное объяснение каждой части кода для варианта 1:

1. `year = int(input("Введите год для проверки: "))` - Эта строка кода просит пользователя ввести год и сохраняет его в переменной `year`.

2. `if year < 1582:` - Этот условный оператор проверяет, меньше ли введенный год 1582. Если это так, то выполняется следующий блок кода.

3. `print(f"Год {year} не соответствует Григорианскому календарю.")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год не соответствует Григорианскому календарю.

4. `elif year % 400 == 0:` - Этот условный оператор проверяет, делится ли введенный год на 400 без остатка. Если это так, то год считается високосным.

5. `print(f"Год {year} является високосным.")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год является високосным.

6. `elif year % 100 == 0:` - Этот условный оператор проверяет, делится ли введенный год на 100 без остатка. Если это так, то год не считается високосным.

7. `print(f"Год {year} не является високосным.")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год не является високосным.

8. `elif year % 4 == 0:` - Этот условный оператор проверяет, делится ли введенный год на 4 без остатка. Если это так, то выполняется следующий блок кода.

9. `print(f"Год {year} является високосным.")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год является високосным.

10. `else:` - Этот блок кода выполняется, если ни одно из предыдущих условий не выполнено, что означает, что введенный год не является високосным.

11. `print(f"Год {year} не является високосным.")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год не является високосным.

Этот код позволяет определить статус года, основываясь на правилах високосных лет, которые включают в себя деление на 4, 100 и 400.


```python
# Вариант 2:

year = int(input("Введите год для проверки: "))

# Високосный год в Григорианском календаре начинается с 1582
if year < 1582:
    message = f"Год {year} не соответствует Григорианскому календарю."
elif (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    message = f"Год {year} является високосным."
else:
    message = f"Год {year} не является високосным."

print(message)
```

5. `elif (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):` - Этот условный оператор проверяет два условия: первое - делится ли введенный год на 400 без остатка, второе - делится ли год на 4 без остатка и не делится на 100 без остатка. Этот код позволяет определить статус года, основываясь на правилах високосных лет, которые включают в себя деление на 4, 100 и 400, а также учитывает переход на Григорианский календарь.


## Задача 2

Напишите программу,которая запрашивает год и проверяет его на високосность.

Распишите все возможные проверки в цепочке `elif`

Не учитывая ввода года по Григорианскому календарю

### Решение задачи 2

```python
# Вариант 1:

# Запрашиваем год у пользователя
year = int(input("Введите год: "))

# Проверяем, является ли год високосным
if year % 4 == 0:
    # Год високосный, если он кратен 4
    if year % 100 == 0:
        # Но не високосный, если кратен 100 и не кратен 400
        if year % 400 != 0:
            print(f"{year} - невисокосный год")
        else:
            print(f"{year} - високосный год")
    else:
        print(f"{year} - високосный год")
else:
    print(f"{year} - невисокосный год")
```

## Этот код используется для определения, является ли введенный год високосным или нет. Вот подробное объяснение каждой части кода для варианта 1 задачи 2:

1. `year = int(input("Введите год: "))` - Эта строка кода просит пользователя ввести год и сохраняет его в переменной `year`.

2. `if year % 4 == 0:` - Это условный оператор, который проверяет, делится ли введенный год на 4 без остатка. Если это так, то выполняется следующий блок кода.

3. `if year % 100 == 0:` - Этот условный оператор проверяет, делится ли введенный год на 100 без остатка. Если это так, то выполняется следующий блок кода.

4. `if year % 400 != 0:` - Этот условный оператор проверяет, не делится ли введенный год на 400 без остатка. Если это так, то год считается невисокосным.

5. `print(f"{year} - невисокосный год")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год является невисокосным.

6. `else:` - Это ключевое слово используется для обозначения альтернативного блока кода, который будет выполнен, если предыдущее условие не выполнено.

7. `print(f"{year} - високосный год")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год является високосным.

8. `else:` - Этот блок кода выполняется, если ни одно из предыдущих условий не выполнено, что означает, что введенный год не является високосным.

9. `print(f"{year} - невисокосный год")` - Эта строка выводит сообщение, информирующее пользователя о том, что введенный год не является високосным.

Этот код позволяет определить статус года, основываясь на правилах високосных лет, которые включают в себя деление на 4, 100 и 400.


```python
# Вариант 2:

year = int(input("Введите год: "))
print(f"{year} - {('високосный' if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) else 'невисокосный')} год")
```

## Этот код используется для определения, является ли введенный год високосным или нет. Вот подробное объяснение каждой части кода для варианта 2 задачи 2:

1. `year = int(input("Введите год: "))` - Эта строка кода просит пользователя ввести год и сохраняет его в переменной `year`.

2. `print(f"{year} - {('високосный' if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) else 'невисокосный')} год")` - Эта строка выполняет две вещи:
   - Выводит сообщение с введенным годом и его статусом (високосный или невисокосный).
   - Использует условный оператор `if` для определения статуса года.

   Условный оператор `if` в данном случае работает следующим образом:
   - Сначала проверяется, делится ли год на 4 без остатка (`year % 4 == 0`). Если это так, то условие истинно.
   - Затем проверяется, делится ли год на 100 без остатка (`year % 100 != 0`). Если это так, то условие ложно, и год не может быть високосным. Однако, если год делится на 100 с остатком, то проверяется, делится ли он также на 400 без остатка (`year % 400 == 0`). Если это так, то год високосный.
   - Если ни одно из этих условий не выполняется, то год считается невисокосным.

   В конце строки используется форматная строка `f"{year} - {('високосный' if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0) else 'невисокосный')} год"` для форматирования вывода.

Этот код позволяет определить статус года, основываясь на правилах високосных лет, которые включают в себя деление на 4, 100 и 400.
